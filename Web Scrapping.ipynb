```python
# Import Module yang Dibutuhkan
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Response HTML yang diakses
url = "https://www.petsecure.com.au/pet-care/a-guide-to-worldwide-pet-ownership/"
response = requests.get(url)
response.status_code

# Parse & Print HTML dengan Bs4
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify()[:1000])  # Print the first 1000 characters of the parsed HTML

# Dapatkan Data Table dari Bs4 dan jelaskan bagaimana proses pengambilan datanya
# Kita mencari tabel dengan judul "Dog" dan mengambil data dari tabel tersebut
tables = soup.find_all('table')
dog_table = None

for table in tables:
    if "Dog" in table.previous_element.previous_element.get_text():
        dog_table = table
        break

# Dapatkan Data Table Headers dan Table Rows kemudian jadikan satu DataFrame dan print menjadi .csv file
headers = []
for th in dog_table.find_all('th'):
    headers.append(th.get_text().strip())

rows = []
for tr in dog_table.find_all('tr')[1:]:  # Skip the header row
    cells = tr.find_all(['td', 'th'])
    row = [cell.get_text().strip() for cell in cells]
    rows.append(row)

# Buat DataFrame
df = pd.DataFrame(rows, columns=headers)

# Simpan menjadi .csv file
df.to_csv('dog_ownership_data.csv', index=False)
print(df.head())

# Opsional (Cleaning or Data Transformation)
# Contoh sederhana untuk menghapus baris yang mungkin kosong atau tidak relevan
df_cleaned = df.dropna()
print(df_cleaned.head())
```
